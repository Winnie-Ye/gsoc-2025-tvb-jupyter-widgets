```python
from tvbwidgets.ui.plotly_backend.head_viewer import HeadViewer
import numpy as np
import ipywidgets as widgets
import plotly.io as pio
pio.renderers.default = 'notebook'

n_vertices = 100
brain_vertices = np.random.rand(n_vertices, 3)  # 100 个 3D 坐标点
brain_triangles = np.array([[i, (i+1)%n_vertices, (i+2)%n_vertices] for i in range(n_vertices - 2)])
activation_data = np.random.rand(n_vertices)  # 每个顶点上的数值，用来映射颜色

viewer = HeadViewer()
viewer.add_brain_surface(
    vertices=brain_vertices,
    triangles=brain_triangles,
    name="Cortex",
    scalars=activation_data
)

controls = widgets.VBox()
viewer.add_transparency_slider(controls)
viewer.add_colormap_selector(controls)

from IPython.display import display
display(controls)
viewer.show()
```
